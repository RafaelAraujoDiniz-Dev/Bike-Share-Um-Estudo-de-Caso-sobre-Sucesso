# üìä Carregando os Dados

```python
# Descoberta autom√°tica de CSVs em INPUT_ROOT
from pathlib import Path
import gc

csv_files: list[Path] = []
for p in INPUT_ROOT.glob('**/*.csv'):
    # Ignora arquivos muito pequenos (ex.: metadados); ajuste se necess√°rio
    if p.stat().st_size > 10_000:
        csv_files.append(p)

# Seleciona apenas 12 arquivos (ajuste conforme necessidade)
csv_files = sorted(csv_files)[-12:]  # use [:12] para pegar os 12 primeiros
print(f'üîç Arquivos CSV selecionados: {len(csv_files)}')

# Leitura e concatena√ß√£o (otimizada para mem√≥ria, sem perder linhas)
DATE_COLS = ['started_at', 'ended_at']
DTYPE_MAP = {
    # manter como string (reduz object pesado e inconsist√™ncia entre arquivos)
    'ride_id': 'string',
    'rideable_type': 'string',
    'start_station_name': 'string',
    'start_station_id': 'string',
    'end_station_name': 'string',
    'end_station_id': 'string',
    'member_casual': 'string',
}

row_counts: dict[str, int] = {}
dfs: list[pd.DataFrame] = []

for fp in csv_files:
    try:
        # Tenta ler j√° parseando datas (economiza mem√≥ria vs strings)
        try:
            df_tmp = pd.read_csv(
                fp,
                low_memory=False,
                dtype=DTYPE_MAP,
                parse_dates=DATE_COLS,
            )
        except ValueError:
            # Caso um arquivo n√£o tenha todas as colunas de data
            df_tmp = pd.read_csv(fp, low_memory=False, dtype=DTYPE_MAP)
            for c in DATE_COLS:
                if c in df_tmp.columns:
                    df_tmp[c] = pd.to_datetime(df_tmp[c], errors='coerce')

        dfs.append(df_tmp)
        row_counts[fp.name] = len(df_tmp)
        print(f'  ‚úì {fp.name} - {len(df_tmp):,} linhas')
    except Exception as e:
        print(f'  ‚úó Falha ao ler {fp.name}: {e}')

if not dfs:
    raise RuntimeError('‚ùå Nenhum CSV v√°lido encontrado. Adicione o dataset via Add data.')

raw = pd.concat(dfs, ignore_index=True)

# Checagem: concatena√ß√£o preservou a soma de linhas lidas
sum_rows = int(sum(row_counts.values()))
if len(raw) != sum_rows:
    raise RuntimeError(f"‚ùå Inconsist√™ncia: raw tem {len(raw):,} linhas, mas a soma dos arquivos √© {sum_rows:,}.")

# Liberar intermedi√°rios (reduz pico de RAM)
del dfs
try:
    del df_tmp
except NameError:
    pass
gc.collect()

print(f'\nüì¶ Total de linhas brutas: {len(raw):,}')
print(f'üìã Colunas: {list(raw.columns)}')
print(f'üíæ Tamanho em mem√≥ria: {raw.memory_usage(deep=True).sum() / 1024**2:.2f} MB')

# Visualizar primeiras linhas
raw.head(3)
```

<div style="background-color:#b6fcbd; padding: 20px;">
<h3 style="font-size:24px; font-family:verdana;"><b>Limpeza de Dados e Padroniza√ß√£o:</b></h3>
<p style="font-size:20px; font-family:verdana; line-height: 1.7em">
Agora vamos realizar a limpeza e padroniza√ß√£o dos dados, incluindo:
</p>
<ul style="font-size:20px; font-family:verdana; line-height: 1.7em">
    <li>Padroniza√ß√£o de nomes de colunas</li>
    <li>Convers√£o de tipos de dados apropriados</li>
    <li>Remo√ß√£o de duplicatas</li>
    <li>Tratamento de valores ausentes</li>
    <li>Valida√ß√£o de dados (datas, coordenadas, etc.)</li>
    <li>Cria√ß√£o de features derivadas</li>
</ul>
</div>

```python
import re
import gc

print('üßπ Iniciando processo de limpeza...\n')

# 1. Padronizar nomes de colunas
raw.columns = [re.sub(r"\s+", "_", c.strip().lower()) for c in raw.columns]
print('‚úì Colunas padronizadas')

# 2. Garantir colunas essenciais
colunas_essenciais = ['ride_id', 'rideable_type', 'started_at', 'ended_at', 'member_casual']
for col in colunas_essenciais:
    if col not in raw.columns:
        raw[col] = pd.NA
print('‚úì Colunas essenciais verificadas')

# 3. Converter tipos de dados
# Datas (se j√° vierem como datetime, isso √© no-op)
for c in ['started_at', 'ended_at']:
    if c in raw.columns and not pd.api.types.is_datetime64_any_dtype(raw[c]):
        raw[c] = pd.to_datetime(raw[c], errors='coerce')
print('‚úì Datas convertidas')

# Normaliza√ß√£o de texto (preserva NA; evita transformar NA em "nan")
colunas_texto = ['rideable_type', 'start_station_name', 'end_station_name', 'member_casual']
for c in colunas_texto:
    if c in raw.columns:
        raw[c] = (
            raw[c]
            .astype('string')
            .str.replace(r"\s+", " ", regex=True)
            .str.strip()
            .str.lower()
        )
print('‚úì Texto normalizado')

# IDs para string (StringDtype preserva NA)
for c in ['ride_id', 'start_station_id', 'end_station_id']:
    if c in raw.columns:
        raw[c] = raw[c].astype('string')
print('‚úì IDs convertidos para string')

# Categorias de baixa cardinalidade (economiza mem√≥ria)
for c in ['member_casual', 'rideable_type']:
    if c in raw.columns:
        raw[c] = raw[c].astype('category')
print('‚úì Colunas categ√≥ricas otimizadas')

# 4. Remover duplicatas (regra existente: por ride_id)
antes_duplicatas = len(raw)
raw = raw.drop_duplicates(subset=['ride_id'])
depois_duplicatas = len(raw)
print(f'‚úì Duplicatas removidas: {antes_duplicatas - depois_duplicatas:,}')

gc.collect()

print(f'\nüì¶ Total de linhas ap√≥s limpeza inicial: {len(raw):,}')
print(f'üíæ Tamanho em mem√≥ria (p√≥s-limpeza): {raw.memory_usage(deep=True).sum() / 1024**2:.2f} MB')
raw.head(3)
```

<div style="background-color:#b6fcbd; padding: 20px;">
<h3 style="font-size:24px; font-family:verdana;"><b>Valida√ß√£o e Filtragem de Dados:</b></h3>
<p style="font-size:20px; font-family:verdana; line-height: 1.7em">
Vamos aplicar regras de neg√≥cio e valida√ß√µes para garantir a qualidade dos dados:
</p>
<ul style="font-size:20px; font-family:verdana; line-height: 1.7em">
    <li><b>Viagens inv√°lidas:</b> Datas ausentes ou inconsistentes (ended_at < started_at)</li>
    <li><b>Dura√ß√£o:</b> Viagens com dura√ß√£o ‚â§ 0 ou > 12 horas (720 minutos)</li>
    <li><b>Tipo de usu√°rio:</b> Apenas 'member' ou 'casual'</li>
    <li><b>Coordenadas:</b> Latitude entre -90¬∞ e 90¬∞, longitude entre -180¬∞ e 180¬∞</li>
    <li><b>Operacional:</b> Viagens muito curtas (< 1 minuto) podem ser testes</li>
</ul>
</div>

```python
import gc

print('üîç Aplicando valida√ß√µes e filtros...\n')

# Calcular dura√ß√£o da viagem SEM criar coluna em raw (reduz RAM se voc√™ mantiver raw)
ride_length_min = (raw['ended_at'] - raw['started_at']).dt.total_seconds() / 60

# M√°scaras de valida√ß√£o
# 1. Datas inv√°lidas
m_invalid_dt = raw['started_at'].isna() | raw['ended_at'].isna() | (raw['ended_at'] < raw['started_at'])
print(f'‚ùå Datas inv√°lidas: {m_invalid_dt.sum():,}')

# 2. Dura√ß√£o inv√°lida
m_le_zero = ride_length_min <= 0
m_gt_max = ride_length_min > 720  # 12 horas
print(f'‚ùå Dura√ß√£o ‚â§ 0: {m_le_zero.sum():,}')
print(f'‚ùå Dura√ß√£o > 12h: {m_gt_max.sum():,}')

# 3. Tipo de usu√°rio inv√°lido
m_member_inv = ~raw['member_casual'].isin(['member', 'casual'])
print(f'‚ùå Tipo de usu√°rio inv√°lido: {m_member_inv.sum():,}')

# 4. Viagens operacionais suspeitas
MIN_OPERATIONAL_MIN = 1
m_ops_short = ride_length_min < MIN_OPERATIONAL_MIN
print(f'‚ùå Viagens < 1 min (suspeitas): {m_ops_short.sum():,}')

# 5. Mesma esta√ß√£o, viagem curta
SAME_STATION_SHORT_MIN = 2
m_same_station_short = False
if {'start_station_id', 'end_station_id'}.issubset(raw.columns):
    m_same_station_short = (
        (raw['start_station_id'] == raw['end_station_id']) &
        (ride_length_min < SAME_STATION_SHORT_MIN)
    )
    print(f'‚ùå Mesma esta√ß√£o + < 2 min: {m_same_station_short.sum() if isinstance(m_same_station_short, pd.Series) else 0:,}')

# 6. Coordenadas inv√°lidas
coord_cols = {'start_lat', 'start_lng', 'end_lat', 'end_lng'}
if coord_cols.issubset(raw.columns):
    m_coords_inv = (
        raw['start_lat'].isna() | raw['start_lng'].isna() |
        raw['end_lat'].isna() | raw['end_lng'].isna() |
        (raw['start_lat'] < -90) | (raw['start_lat'] > 90) |
        (raw['end_lat'] < -90) | (raw['end_lat'] > 90) |
        (raw['start_lng'] < -180) | (raw['start_lng'] > 180) |
        (raw['end_lng'] < -180) | (raw['end_lng'] > 180)
    )
    print(f'‚ùå Coordenadas inv√°lidas: {m_coords_inv.sum():,}')
else:
    m_coords_inv = False

# Filtro combinado
invalid = m_invalid_dt | m_le_zero | m_gt_max | m_member_inv | m_ops_short
if isinstance(m_same_station_short, pd.Series):
    invalid = invalid | m_same_station_short
if isinstance(m_coords_inv, pd.Series):
    invalid = invalid | m_coords_inv

# Criar dataset limpo
clean = raw.loc[~invalid].copy()
# Garantir que ride_length_min exista em clean (sem inchar raw)
clean['ride_length_min'] = ride_length_min.loc[~invalid].to_numpy()

# Otimiza√ß√£o de mem√≥ria em clean (baixa cardinalidade)
for c in ['member_casual', 'rideable_type']:
    if c in clean.columns:
        clean[c] = clean[c].astype('category')

print(f'\nüìä Resumo da filtragem:')
print(f'  ‚Ä¢ Registros originais: {len(raw):,}')
print(f'  ‚Ä¢ Registros inv√°lidos: {invalid.sum():,}')
print(f'  ‚Ä¢ Registros v√°lidos: {len(clean):,}')
print(f'  ‚Ä¢ Taxa de reten√ß√£o: {len(clean)/len(raw)*100:.2f}%')

# Liberar grandes intermedi√°rios booleanos/series
for _name in [
    'ride_length_min', 'invalid', 'm_invalid_dt', 'm_le_zero', 'm_gt_max',
    'm_member_inv', 'm_ops_short', 'm_same_station_short', 'm_coords_inv'
]:
    if _name in locals():
        del locals()[_name]

gc.collect()

print(f'üíæ Mem√≥ria estimada: raw={raw.memory_usage(deep=True).sum()/1024**2:.2f} MB | clean={clean.memory_usage(deep=True).sum()/1024**2:.2f} MB')
clean.head(3)
```

<div style="background-color:#b6fcbd; padding: 20px;">
<h3 style="font-size:24px; font-family:verdana;"><b>Engenharia de Features:</b></h3>
<p style="font-size:20px; font-family:verdana; line-height: 1.7em">
Vamos criar features adicionais para enriquecer nossa an√°lise:
</p>
<ul style="font-size:20px; font-family:verdana; line-height: 1.7em">
    <li><b>Temporais:</b> Dia da semana, hora do dia, m√™s, ano, esta√ß√£o do ano</li>
    <li><b>Categ√≥ricas:</b> Tipo de dia (√∫til/fim de semana), per√≠odo do dia (manh√£/tarde/noite)</li>
    <li><b>Geogr√°ficas:</b> Dist√¢ncia percorrida (usando f√≥rmula de Haversine)</li>
    <li><b>Comportamentais:</b> Categorias de dura√ß√£o e dist√¢ncia</li>
</ul>
</div>

```python
print('üõ†Ô∏è Criando features derivadas...\n')

# Garantia de execu√ß√£o (evita NameError se voc√™ rodar as c√©lulas fora de ordem)
if 'clean' not in globals():
    if 'raw' not in globals():
        raise NameError("'clean' n√£o existe e 'raw' n√£o foi carregado. Execute a c√©lula de carregamento primeiro.")
    clean = raw.copy()
    print("‚ÑπÔ∏è 'clean' n√£o estava definido; usando clean = raw.copy()")

# Garantir datetime
for c in ['started_at', 'ended_at']:
    if c in clean.columns and not pd.api.types.is_datetime64_any_dtype(clean[c]):
        clean[c] = pd.to_datetime(clean[c], errors='coerce')

# 1. Features temporais
clean['weekday_idx'] = clean['started_at'].dt.weekday.astype('int8')
clean['day_of_week'] = clean['started_at'].dt.day_name()
clean['day_of_week_pt'] = clean['weekday_idx'].map({
    0: 'Segunda', 1: 'Ter√ßa', 2: 'Quarta', 3: 'Quinta',
    4: 'Sexta', 5: 'S√°bado', 6: 'Domingo'
})
clean['month'] = clean['started_at'].dt.month
clean['month_name'] = clean['started_at'].dt.month_name()
clean['year'] = clean['started_at'].dt.year
clean['hour_of_day'] = clean['started_at'].dt.hour.astype('uint8')
clean['date'] = clean['started_at'].dt.date
print('‚úì Features temporais criadas')

# 2. Tipo de dia
clean['is_weekend'] = clean['weekday_idx'].isin([5, 6])
clean['day_type'] = clean['is_weekend'].map({True: 'Fim de Semana', False: 'Dia √ötil'})
print('‚úì Tipo de dia criado')

# 3. Per√≠odo do dia
def get_period(hour):
    if 5 <= hour < 12:
        return 'Manh√£'
    elif 12 <= hour < 18:
        return 'Tarde'
    elif 18 <= hour < 22:
        return 'Noite'
    else:
        return 'Madrugada'

clean['period_of_day'] = clean['hour_of_day'].apply(get_period)
print('‚úì Per√≠odo do dia criado')

# 4. Esta√ß√£o do ano (hemisf√©rio norte)
def get_season(month):
    if month in [12, 1, 2]:
        return 'Inverno'
    elif month in [3, 4, 5]:
        return 'Primavera'
    elif month in [6, 7, 8]:
        return 'Ver√£o'
    else:
        return 'Outono'

clean['season'] = clean['month'].apply(get_season)
print('‚úì Esta√ß√£o do ano criada')

# 5. Dist√¢ncia geogr√°fica (Haversine) ‚Äî vetorizado (muito mais r√°pido que apply)
coord_cols = {'start_lat', 'start_lng', 'end_lat', 'end_lng'}
if coord_cols.issubset(clean.columns):
    mask = (
        clean['start_lat'].notna() & clean['start_lng'].notna() &
        clean['end_lat'].notna() & clean['end_lng'].notna()
    )
    # Inicializa coluna com NaN
    clean['distance_km'] = np.nan
    if mask.any():
        R = 6371.0  # km
        lat1 = np.radians(clean.loc[mask, 'start_lat'].to_numpy(dtype='float64'))
        lon1 = np.radians(clean.loc[mask, 'start_lng'].to_numpy(dtype='float64'))
        lat2 = np.radians(clean.loc[mask, 'end_lat'].to_numpy(dtype='float64'))
        lon2 = np.radians(clean.loc[mask, 'end_lng'].to_numpy(dtype='float64'))
        dlat = lat2 - lat1
        dlon = lon2 - lon1
        a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * (np.sin(dlon / 2) ** 2)
        c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
        dist = R * c
        clean.loc[mask, 'distance_km'] = dist
        # reduzir mem√≥ria
        clean['distance_km'] = clean['distance_km'].astype('float32')
    print('‚úì Dist√¢ncia calculada (vetorizado)')
else:
    clean['distance_km'] = np.nan
    print('‚ö†Ô∏è  Colunas de coordenadas n√£o encontradas - dist√¢ncia n√£o calculada')

# 6. Categorias de dura√ß√£o
bins_duration = [0, 5, 10, 20, 30, 60, np.inf]
labels_duration = ['<5 min', '5-10 min', '10-20 min', '20-30 min', '30-60 min', '60+ min']
clean['duration_category'] = pd.cut(clean['ride_length_min'], bins=bins_duration, labels=labels_duration)
print('‚úì Categorias de dura√ß√£o criadas')

# 7. Categorias de dist√¢ncia
if 'distance_km' in clean.columns and clean['distance_km'].notna().sum() > 0:
    bins_distance = [0, 1, 2, 3, 5, 10, np.inf]
    labels_distance = ['<1 km', '1-2 km', '2-3 km', '3-5 km', '5-10 km', '10+ km']
    clean['distance_category'] = pd.cut(clean['distance_km'], bins=bins_distance, labels=labels_distance)
    print('‚úì Categorias de dist√¢ncia criadas')

print(f'\nüìä Dataset final: {clean.shape}')
print(f'üìã Total de features: {len(clean.columns)}')
print(f'\nüéØ Amostra do dataset limpo:')
clean.head(3)
```

## üìä An√°lise Explorat√≥ria de Dados (EDA)

```python
# KPIs principais
print('='*80)
print('üìä INDICADORES-CHAVE DE PERFORMANCE (KPIs)')
print('='*80)

total_viagens = len(clean)
print(f'\nüö¥ Total de viagens: {total_viagens:,}')

# Distribui√ß√£o por tipo de usu√°rio
user_counts = clean['member_casual'].value_counts()
user_pct = clean['member_casual'].value_counts(normalize=True) * 100

print(f'\nüë• Distribui√ß√£o por tipo de usu√°rio:')
for user_type in ['member', 'casual']:
    if user_type in user_counts.index:
        count = user_counts.get(user_type, 0)
        pct = user_pct.get(user_type, 0)
        print(f'  ‚Ä¢ {user_type.title()}: {count:,} ({pct:.2f}%)')

# Estat√≠sticas de dura√ß√£o
print(f'\n‚è±Ô∏è  Estat√≠sticas de dura√ß√£o (em minutos):')
print(f'  ‚Ä¢ M√©dia: {clean["ride_length_min"].mean():.2f}')
print(f'  ‚Ä¢ Mediana: {clean["ride_length_min"].median():.2f}')
print(f'  ‚Ä¢ M√≠nimo: {clean["ride_length_min"].min():.2f}')
print(f'  ‚Ä¢ M√°ximo: {clean["ride_length_min"].max():.2f}')
print(f'  ‚Ä¢ Desvio padr√£o: {clean["ride_length_min"].std():.2f}')

# Estat√≠sticas por tipo de usu√°rio
print(f'\n‚è±Ô∏è  Dura√ß√£o m√©dia por tipo de usu√°rio:')
for user_type in ['member', 'casual']:
    if user_type in clean['member_casual'].values:
        avg_duration = clean[clean['member_casual'] == user_type]['ride_length_min'].mean()
        print(f'  ‚Ä¢ {user_type.title()}: {avg_duration:.2f} min')

# Estat√≠sticas de dist√¢ncia (se dispon√≠vel)
if 'distance_km' in clean.columns and clean['distance_km'].notna().sum() > 0:
    print(f'\nüìè Estat√≠sticas de dist√¢ncia (em km):')
    print(f'  ‚Ä¢ M√©dia: {clean["distance_km"].mean():.2f}')
    print(f'  ‚Ä¢ Mediana: {clean["distance_km"].median():.2f}')
    print(f'  ‚Ä¢ M√≠nimo: {clean["distance_km"].min():.2f}')
    print(f'  ‚Ä¢ M√°ximo: {clean["distance_km"].max():.2f}')

# Distribui√ß√£o por tipo de bicicleta
print(f'\nüö≤ Distribui√ß√£o por tipo de bicicleta:')
bike_counts = clean['rideable_type'].value_counts()
for bike_type, count in bike_counts.items():
    pct = count / total_viagens * 100
    print(f'  ‚Ä¢ {bike_type}: {count:,} ({pct:.2f}%)')

print(f'\n' + '='*80)
```

```python
# 5) Top 10 Esta√ß√µes Iniciais (barras horizontais)
if 'start_station_name' not in clean.columns:
    raise KeyError("Coluna 'start_station_name' n√£o encontrada no dataset.")

top10_start = (
    clean['start_station_name']
    .dropna()
    .astype('string')
    .str.strip()
    .replace({'': pd.NA})
    .dropna()
    .value_counts()
    .head(10)
    .sort_values(ascending=True)
    .rename_axis('start_station_name')
    .reset_index(name='trips')
 )

fig_top10 = px.bar(
    top10_start,
    x='trips',
    y='start_station_name',
    orientation='h',
    title='<b>Top 10 Esta√ß√µes Iniciais</b>',
    labels={'start_station_name': 'Esta√ß√£o inicial', 'trips': 'N√∫mero de viagens'},
    template='plotly_white',
 )
fig_top10.update_traces(hovertemplate='Esta√ß√£o: %{y}<br>Viagens: %{x:,}<extra></extra>')
fig_top10.update_layout(height=500)
display_plotly(fig_top10)
```
